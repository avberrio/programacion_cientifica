## Preliminares

Importamos las librerías que usaremos durante el desarrollo de la actividad

````python
%load_ext memory_profiler
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
import h5py

def image_plot(img):
    """
    img is an (nx,ny,3) numpy.array
    """
    plt.figure(figsize=(12,12))
    plt.imshow(img)
    plt.axis('off')
    plt.show()
````

Importamos y mostramos la imagen que usaremos durante la experiencia

````python
img = plt.imread('heic1608b.jpg', format='jpeg')[0:8660,:,:]
image_plot(img)
````

## Paso 1

Se inicia creando el archivo `HDF5`:

````python
file = h5py.File('stars_hdf5.h5', 'w')
````

Luego, se procede a crear el dataset que contendrá matriz que representa la imagen:

````python
dataset = file.create_dataset("stars_dataset", img.shape, dtype='uint8', chunks = (100, 100, 1))
````

Importante destacar, que para el desarrollo de esta experiencia se utilizarán *chunks* como formato para almacenar el archivo `HDF5`, dado que será bastante conveniente al momento de aplicar el kernel durante el ítem 1 del paso 2. Notar que se usó un chunk de $20 \times 20$ porque al dividir la cantidad de filas y columnas por este valor, se obtiene un entero.

Se guarda la imagen en el dataset y de paso se borra `img`:

````python
for k in range(3):
    for j in range(img.shape[0]):
        dataset[j, :, k] = img[j, :, k]

del img
````

## Paso 2

Se define la función que aplica el kernel

```python
K = 1/25. * np.ones((5,5))
```

sin overlap a sólo 1 canal de la imagen:

````python
def apply_filter(dataset, channel):
    filtered_image = np.empty((dataset.shape[0], dataset.shape[1]), dtype='uint8')
    for i in range(0, dataset.shape[0], 5):
        for j in range(0, dataset.shape[1], 5):
            filtered_image[i + 2, j + 2] = 1/25. * np.sum(dataset[i : i + 5, j : j + 5, channel])
    return filtered_image
````

Nótese que la función `apply_filter` no carga la toda la imagen almacenada en `dataset` en memoria, solo unos pequeños pedacitos.

Se usa la función `apply_filter` para obtener la matriz que representa sólo 1 canal de la imagen tras aplicar el kernel `K`. De paso, se realizan `timeit` y `memit` para los computos asociados:

````python
%%timeit -r 1
%%memit
matrix_a = apply_filter(dataset, 0)
````

````python
%%timeit -r 1
%%memit
matrix_b = apply_filter(dataset, 1)
````

````python
%%timeit -r 1
%%memit
matrix_c = apply_filter(dataset, 2)
````

Luego, se procede a ver cada canal por separado de la imagen generada tras aplicar el kernel a la imagen original:

````python
image_plot(matrix_a)
````

````python
image_plot(matrix_b)
````

````python
image_plot(matrix_c)
````

Las imágenes generadas anteriormente, describen mas o menos cómo debería quedar la imagen (con los tres canales) tras aplicar el filtro. Para ello, se construye la imagen final con los tres canales:

````python
result_image = np.empty(dataset.shape, dtype='uint8')
for j in range(dataset.shape[1]):
    result_image[:, j, 0] = matrix_a[:, j]
    result_image[:, j, 1] = matrix_b[:, j]
    result_image[:, j, 2] = matrix_c[:, j]
````

y se muestra por pantalla:

````python
image_plot(result_image)
````

El kernel aplicado a la imagen original lo que hizo fue crear una nueva imagen, tal que cada 5 pixeles de separación existe otro pixel cuyo color está dado por el promedio de colores de la vecindad de una matriz de orden $5\times 5$ del punto en cuestión.

Por otro lado, la separación de 5 pixeles para llegar a "uno de color" se debe a que el kernel fue aplicado sin overlap.

## Extras

Si eliminamos la separación entre 5 pixeles de la imagen generada tras aplicar el kernel `K`, la imagen resultante es una versión comprimida de la original. Para demostrarlo, creamos una matriz que elimine la separación de 5 pixeles a través de:

````python
matrix_final = np.empty((int(dataset.shape[0]/5), int(dataset.shape[1]/5), dataset.shape[2]), dtype='uint8')
for i in range(int(dataset.shape[0]/5)):
    for j in range(int(dataset.shape[1]/5)):
        matrix_final[i, j, :] = result_image[i * 5 + 2, j * 5 + 2, :]
````

Luego, mostramos la imagen por pantalla:

````python
image_plot(matrix_final)
````
