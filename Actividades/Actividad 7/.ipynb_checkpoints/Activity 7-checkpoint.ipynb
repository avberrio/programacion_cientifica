## 2.2. &nbsp;&nbsp;&nbsp; Pregunta 2

Graficamos la función $f(z) = \tanh(z^2)$ que define el comportamiento del sistema en el intervalo $\left[0, 1\right]$:

```python
z = np.linspace(0, 1, n)
plt.plot(z, np.tanh(z ** 2))
plt.show()
```

Realizamos otro gráfico de la función anterior, pero esta vez, con dominio $z\in \left[ -2, 2\right]$ para observar globalmente el comportamiento de la función:

```python
z = np.linspace(-2, 2, n)
plt.plot(z, np.tanh(z ** 2))
plt.show()
```

## 2.3. &nbsp;&nbsp;&nbsp; Pregunta 3

Hacemos la función `simular(x0, y0, dt, tf)` que recibe los `np.ndarray` con las coordenadas iniciales `x0` e `y0` del sistema, el _time step_ `dt` y el tiempo final de simulación:

```python
def simular(x0, y0, dt, tf):
    xi = x0
    yi = y0
    for i in range(tf):
        xi1 = xi + dt * np.tanh(xi ** 2 + yi ** 2)
        yi1 = yi + dt * np.tanh(xi ** 2 + yi ** 2)
        xi = xi1
        yi = yi1
    plt.plot(xi, yi, '.')
    plt.show()
```

```python
simular(x, y, 0.01, 4)
```

## 2.4. &nbsp;&nbsp;&nbsp; Pregunta 4

Usamos `interact` para desarrollar un _widget_ que nos permite visualizar interactivamente distintos estados finales del sistema (distintos `tf`). Para ello, dejaremos fijos los parámetros `x0, y0, dt` en `simular`:

```python
interact(simular, x0 = fixed(x), y0 = fixed(y), dt = fixed(0.01), tf = (1, 500, 1))
```
