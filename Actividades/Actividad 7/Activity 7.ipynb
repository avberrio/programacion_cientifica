## 2.2. &nbsp;&nbsp;&nbsp; Pregunta 2

Graficamos la función $f(z) = \tanh(z^2)$ que define el comportamiento del sistema en el intervalo $\left[0, 1\right]$:

```python
z = np.linspace(0, 1, n)
plt.plot(z, np.tanh(z ** 2))
plt.show()
```

Realizamos otro gráfico de la función anterior, pero esta vez, con dominio $z\in \left[ -2, 2\right]$ para observar globalmente el comportamiento de la función:

```python
z = np.linspace(-2, 2, n)
plt.plot(z, np.tanh(z ** 2))
plt.show()
```
